# TASK 3
Create a bar plot of the total number of homeless people by region in the `homelessness` DataFrame. Order the bars in descending order. Bonus: create a horizontal bar chart.

```python
homelessness = pd.read_csv('/Users/julio/Desktop/DATA/PROJECTS/DATACAMP/Data Manipulation with pandas/homelessness.csv')
```

```python
#homelessness dataset
homelessness.head()
homelessness = homelessness.drop(columns='Unnamed: 0')
homelessness.head()
```

```python
homelessness_region = homelessness.groupby('region').agg({'individuals':'sum'})
homelessness_region
```

```python
homelessness_region.sort_values('individuals', ascending=True).plot(kind='barh', color='#f4acb7').get_legend().remove()
plt.title = 'Total number of homeless people by region'
plt.xlabel('individuals')
plt.show()
```

# TASK 4
Create a line plot with two lines representing the temperatures in Toronto and Rome. Make sure to properly label your plot. Bonus: add a legend for the two lines.

```python
#Temperatures
temperatures = pd.read_csv('temperatures.csv', parse_dates=['date'])
temperatures = temperatures.drop(columns='Unnamed: 0')
temperatures.head()
```

```python
type(temperatures.date[0])
```

```python
temp_Rome_Toronto = temperatures[temperatures['city'].isin(['Toronto', 'Rome'])]
temp_Rome_Toronto.city.unique()
```

```python
temp_Rome = temperatures[temperatures['city']=='Rome']
temp_Toronto = temperatures[temperatures['city']=='Toronto']
plt.plot(temp_Rome.date, temp_Rome.avg_temp_c, label='Rome')
plt.plot(temp_Toronto.date, temp_Toronto.avg_temp_c, label='Toronto')
plt.xlabel('dates')
plt.ylabel('temperature')
plt.legend(loc='best', fontsize = 'small')
plt.show()
```
